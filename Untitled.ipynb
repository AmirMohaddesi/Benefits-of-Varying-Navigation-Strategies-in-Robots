```python
import seaborn as sns

import matplotlib as mpl
mpl.rcParams['font.family'] = 'Arial'
mpl.rcParams['pdf.fonttype'] = 42

import matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
map = pd.read_csv("SpikeWave/edited map2.tsv", sep='\t', header=None)
goals = pd.read_csv("Goals.txt", sep='\t', header=None)
map[map <= 399] = 0.1
map[map > 399] = 1
for i in range(goals.shape[0]):
    map.iloc[goals.iloc[i, 0], goals.iloc[i, 1]] = 0.5
map
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, 
                        figsize=(12,10), 
                        gridspec_kw={'width_ratios': [10, 2]})

sns.heatmap(map.T, cmap="flare",
            linewidths=0.01, linecolor='gray',
            cbar=False, 
            ax=axs[0])

axs[0].set(xticklabels=[])
axs[0].set(yticklabels=[])
axs[0].tick_params(bottom=False, left=False)

cmap = matplotlib.cm.get_cmap('flare')
handles = []
patch = mpatches.Patch(color=cmap(0), label='Hallways')
handles.append(patch)

patch = mpatches.Patch(color=cmap(128), label='Goal locations')
handles.append(patch)

patch = mpatches.Patch(color=cmap(255), label='Unpassable terrain')
handles.append(patch)

axs[1].legend(loc='center left',
              handles=handles,
              fontsize=20,
              title_fontsize=30)
axs[1].axis('off')

plt.tight_layout()
plt.savefig("figure2.pdf")
```

```python
map = pd.read_csv("SpikeWave/edited map2.tsv", sep='\t', header=None)
map[map == 4] = 40
map
```
